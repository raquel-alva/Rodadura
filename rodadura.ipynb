## Rodadura
Un sólido rígido cumple la condición de rodadura si gira sin deslizar a la vez que se traslada


```python
# RODADURA
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation,rc
from IPython.display import HTML

fig,ax = plt.subplots()
ax.set_xlim((-1,7))
ax.set_ylim((-2,2))
ax.set_aspect('equal')
fig.set_size_inches(14, 5)

ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

line,=ax.plot([],[],linewidth=2)
line2,=ax.plot([],[],linewidth=2,color='magenta')
line3,=ax.plot([],[],linewidth=2,marker='o')
line4,=ax.plot([],[],linewidth=2,marker='o')
line5,=ax.plot([],[],linewidth=2,marker='o')

angu = np.linspace( 0 , 2 * np.pi , 100 )
angu2= np.linspace( 0,  np.pi/2, 100)
radi = np.sqrt(1)

ax.plot([0,np.pi/2],[-1,-1],'m-')


def init():
    line.set_data([],[])
    line2.set_data([],[])
    line3.set_data([],[])
    line4.set_data([],[])
    line5.set_data([],[])

    return(line,line2,line3,line4,line5)

cero = np.zeros(100)


def animate(i):
    x=radi * np.cos( angu )+i/40
    x0 = radi*np.cos(angu/4+i/40)+i/40
    x00 = radi*np.cos(i/40)+i/40
    y=radi * np.sin( angu )
    y0 = -radi*np.sin(angu/4+i/40)
    y00 = -radi*np.sin(i/40)
    line.set_data(x,y)
    x000 = cero +i/40
    y000 = cero 
    xx0=radi*np.cos(i/40+np.pi/2)+i/40
    yy0=-radi*np.sin(i/40+np.pi/2)
    line2.set_data(x0,y0)
    line3.set_data(x00,y00)
    line4.set_data(x000,y000)
    line5.set_data(xx0,yy0)

    return(line,line2,line3,line4,line5)


anim=animation.FuncAnimation(fig,animate,init_func=init,frames=55,interval=100,blit=True)

#plt.show()

HTML(anim.to_html5_video())
#anim.save('rodadura.mp4', fps=30)
```

## Rotación con deslizamiento
No se cumple la condición de rodadura si el punto de contacto entre el sólido y la superficie sobre la que
rueda, tiene una velocidad distinta de 0, ya sea hacia delante o hacia atrás. Un ejemplo de velocidad del punto de apoyo hacia delante sucedería cuando un coche circula sobre una pista con hielo (resbala). La velocidad del punto de apoyo hacia atrás tendría lugar para un coche que circule sobre una pista con grava (derrapa).

```python
# ROTACIÓN CON DESLIZAMIENTO HACIA DELANTE
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation,rc
from IPython.display import HTML


fig,ax = plt.subplots()
ax.set_xlim((-1,7))
ax.set_ylim((-2,2))
ax.set_aspect('equal')
fig.set_size_inches(14, 5)

ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

line,=ax.plot([],[],linewidth=2)
line2,=ax.plot([],[],linewidth=2,color='magenta')
line3,=ax.plot([],[],linewidth=2,marker='o')
line4,=ax.plot([],[],linewidth=2,marker='o')
line5,=ax.plot([],[],linewidth=2,marker='o')

angu = np.linspace( 0 , 2 * np.pi , 100 )
angu2= np.linspace( 0,  np.pi/2, 100)
radi = np.sqrt(1)

ax.plot([0,np.pi/2],[-1,-1],'m-')
ax.plot([np.pi/2,np.pi],[-1,-1],'c--')


def init():
    line.set_data([],[])
    line2.set_data([],[])
    line3.set_data([],[])
    line4.set_data([],[])
    line5.set_data([],[])

    return(line,line2,line3,line4,line5)

cero = np.zeros(100)


def animate(i):
    x=radi * np.cos( angu )+2*i/40
    x0 = radi*np.cos(angu/4+i/40)+2*i/40
    x00 = radi*np.cos(i/40)+2*i/40
    y=radi * np.sin( angu )
    y0 = -radi*np.sin(angu/4+i/40)
    y00 = -radi*np.sin(i/40)
    line.set_data(x,y)
    x000 = cero +2*i/40
    y000 = cero 
    line2.set_data(x0,y0)
    line3.set_data(x00,y00)
    line4.set_data(x000,y000)
    xx0=radi*np.cos(i/40+np.pi/2)+2*i/40
    yy0=-radi*np.sin(i/40+np.pi/2)
    line5.set_data(xx0,yy0)

    return(line,line2,line3,line4,line5)


anim=animation.FuncAnimation(fig,animate,init_func=init,frames=60,interval=100,blit=True)

#plt.show()

HTML(anim.to_html5_video())
#anim.save('rodadura2.mp4', fps=30)
```

```python
# ROTACIÓN CON DESLIZAMIENTO HACIA ATRÁS
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation,rc
from IPython.display import HTML


fig,ax = plt.subplots()
ax.set_xlim((-1,7))
ax.set_ylim((-2,2))
ax.set_aspect('equal')
fig.set_size_inches(14, 5)

ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

line,=ax.plot([],[],linewidth=2)
line2,=ax.plot([],[],linewidth=2,color='magenta')
line3,=ax.plot([],[],linewidth=2,marker='o')
line4,=ax.plot([],[],linewidth=2,marker='o')
line5,=ax.plot([],[],linewidth=2,marker='o')

angu = np.linspace( 0 , 2 * np.pi , 100 )
angu2= np.linspace( 0,  np.pi/2, 100)
radi = np.sqrt(1)

ax.plot([0,np.pi/4],[-1,-1],'m-')
ax.plot([np.pi/4,np.pi/2],[-1,-1],'m--')


def init():
    line.set_data([],[])
    line2.set_data([],[])

    return(line,line2,line3)

cero = np.zeros(100)


def animate(i):
    x=radi * np.cos( angu )+0.5*i/40
    x0 = radi*np.cos(angu/4+i/40)+0.5*i/40
    x00 = radi*np.cos(i/40)+0.5*i/40
    y=radi * np.sin( angu )
    y0 = -radi*np.sin(angu/4+i/40)
    y00 = -radi*np.sin(i/40)
    line.set_data(x,y)
    x000 = cero +0.5*i/40
    y000 = cero 
    line2.set_data(x0,y0)
    line3.set_data(x00,y00)
    line4.set_data(x000,y000)
    xx0=radi*np.cos(i/40+np.pi/2)+0.5*i/40
    yy0=-radi*np.sin(i/40+np.pi/2)
    line5.set_data(xx0,yy0)

    return(line,line2,line3)


anim=animation.FuncAnimation(fig,animate,init_func=init,frames=55,interval=100,blit=True)

#plt.show()

HTML(anim.to_html5_video())
#anim.save('rodadura3.mp4', fps=30)
```
